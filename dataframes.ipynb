Copy the code from the lesson to create a dataframe full of student grades.

```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

```python
df
```

Create a column named passing_english that indicates whether each student has a passing grade in english.

```python
passing_english = df['english'] >= 70
df['passing_english'] = passing_english.astype(int)  # Convert boolean values to integers (0 or 1)
print(df)
```

Which dodge car has the best average mileage? The worst?

Filter the 'dodge' manufacturer car that has the best average mileage using python.

import pandas as pd
import numpy as np
from pydataset import data
mpg = data('mpg')

```python
dodges = mpg[mpg.manufacturer == 'dodge']
```

```python
dodges[dodges.average_mileage == dodges.average_mileage.max()]
```

```python
dodges[dodges.average_mileage == dodges.average_mileage.min()]
```

```python
dodge_cars = [['manufacturer'] == 'dodge', 'average_mileage']
dodge_avg_mileage = dodge_cars(['model'])['average_mileage'].mean().reset_index()
best_dodge_model = dodge_avg_mileage.loc[dodge_avg_mileage['average_mileage'].idxmax()]

print(best_dodge_model)
```

```python
# To filter the 'dodge' manufacturer car that has the best average mileage, we can use the following code:

dodge_cars = mpg[mpg['manufacturer'] == 'dodge']
dodge_avg_mileage = dodge_cars.groupby(['model'])['mpg'].mean().reset_index()
best_dodge_model = dodge_avg_mileage.loc[dodge_avg_mileage['mpg'].idxmax()]

print(best_dodge_model)

# This code first filters the 'mpg' dataset to only include cars manufactured by 'dodge'. Then, it groups the cars by model and calculates the average mpg for each model. Finally, it selects the row with the highest average mpg for a dodge car and prints out the model name and mpg.
```
