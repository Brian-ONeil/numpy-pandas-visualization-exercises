```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Use pandas and matplotlib to plot the following equation:

y=x2−x+2

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
# Generate x points
x = range(-10, 11)

# Generate y points
y = [n**2 - n + 2 for n in x]

# Create dataframe
df = pd.DataFrame({'x': x, 'y': y})

# Plot the equation
plt.plot('x', 'y', data=df)

# Add annotation for point (0, 0)
plt.annotate('Origin (0, 0)', xy=(0,0), xytext=(2, 50),
             textcoords='offset points', ha='left', va='top',
             arrowprops=dict(facecolor='black', shrink=0.05))

# Show the plot
plt.show()
```

Using python matplotlib Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

y=√x
y=x3
y=2x
y=1/(x+1)

You can use functions from the math module to help implement some of the equations above.

```python
# Range for x
x = np.linspace(0, 10, 100)

# y = √x
y1 = np.sqrt(x)
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = √x')

# y = x^3
y2 = x**3
plt.figure()
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^3')

# y = 2x
y3 = 2*x
plt.figure()
plt.plot(x, y3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 2x')

# y = 1/(x+1)
y4 = 1/(x+1)
plt.figure()
plt.plot(x, y4)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 1/(x+1)')

plt.show()
```
