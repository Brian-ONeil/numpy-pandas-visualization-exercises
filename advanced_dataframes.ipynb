Copy the users and roles DataFrames from the examples above.

```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

What is the result of using a right join on the DataFrames? You get id, name, role_id, _merge with only the names on the right rows.

```python
pd.merge(users, roles, how ='right', left_on='role_id', right_on='id')
```

What is the result of using an outer join on the DataFrames? It combines names.

```python
pd.merge(users, roles, how = 'outer', indicator = True)
```

What happens if you drop the foreign keys from the DataFrames and try to merge them? Not much, nothing to merge

```python
users = users.drop(columns='role_id')
```

```python
pd.merge(users, roles, how = 'inner', indicator = True)
```
